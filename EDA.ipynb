```python
# checking for null values
df.isnull().sum()
```

```python
# changing incorrect data type of TotalCharges and checking for null values again

df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df.isnull().sum()
```

```python
# TotalCharges null values are handled by replacing them with  the median

df['TotalCharges'].fillna(df['TotalCharges'].median(), inplace=True)
```

```python
df.describe(include="all")
```

```python
# checking the distribution of target variable
df['Churn'].value_counts()
df['Churn'].value_counts(normalize=True) * 100
```

```python
import seaborn as sns

sns.set(style="whitegrid")
```

```python
sns.countplot(data=df, x='Churn')
plt.title("Churn Distribution")
plt.show()
```

```python
df[['tenure','MonthlyCharges','TotalCharges']].hist(bins=30, figsize=(12, 6))
plt.show()
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns

for col in categorical_cols:
    plt.figure(figsize=(8,4))
    sns.countplot(data=df, x=col, hue="Churn")
    plt.title(f'{col} vs Churn')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# drop identifier columns that shouldn't be correlated
df_plot = df.drop(columns=['customerID'], errors='ignore')

# select numeric columns only
numeric_cols = df_plot.select_dtypes(include=['number']).columns.tolist()
print("Numeric columns used for correlation:", numeric_cols)

# compute correlation matrix
corr = df_plot[numeric_cols].corr()

# optional: replace any NaNs in the correlation matrix (can happen if a column is constant)
corr = corr.fillna(0)

# plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, fmt=".2f", vmin=-1, vmax=1, square=True, cmap='Blues')
plt.title("Correlation matrix (numeric features)")
plt.show()
```
